```python
import pandas as pd
```

```python
df = pd.read_csv('data/space_funding.csv')
df
```

```python
restricted_set = df[['Company','Type of transaction','Sector','Country','Amount USD']]
```

```python
restricted_set['amount_preprocessed'] = restricted_set['Amount USD'].str.extract(r'(\d+.?\d)').astype('float')
```

```python
restricted_set['pkey'] = restricted_set.index
```

```python
restricted_set.drop('Amount USD',axis=1,inplace=True)
```

```python
restricted_set.to_csv('data/preprocessed_funding.csv')
```
